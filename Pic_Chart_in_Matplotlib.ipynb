```python
import matplotlib.pyplot as plt
plt.style.use("tableau-colorblind10")
slices = [20,25,30,50,60]
labels = ["Mango", "Orange", "Apple", "Pine", "lyche"]
colors = ["red", "green", "blue", "Orange","purple"]
explode =[0,0,0,0.3,0]

plt.pie(slices, shadow=True, labels = labels, autopct='%1.1f%%',  startangle=90, colors = colors, explode=explode, wedgeprops={'edgecolor':'black'})
plt.title("Slices of Fruits")
plt.show()
```
