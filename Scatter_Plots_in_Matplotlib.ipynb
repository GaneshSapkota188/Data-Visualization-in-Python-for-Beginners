```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,8,7,23,45,46,57,22,35,1,2,6,23]
y = [12,21,4,6,77,8,23,21,43,5,6,9,1,1,2,45,67,46]

c=[6,7,8,4,5,6,7,8,9,9,10,4,5,6,7,7,8,4]

plt.scatter(x,y, c=c, cmap = "Reds", edgecolor="black", linewidth=1, s=150)

plt.title("Data Vis..")
plt.xlabel("x Axis")
plt.ylabel("y Axis")

cbar = plt.colorbar()
cbar.set_label("Darkness")

plt.show()
```
