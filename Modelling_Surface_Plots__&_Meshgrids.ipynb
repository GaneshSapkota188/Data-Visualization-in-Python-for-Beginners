```python
import matplotlib.pyplot as plt
import numpy as np

axis  = plt.axes(projection="3d")

x = np.arange(-2,2,0.02)
y  = np.arange(-2,2,0.02)

meshx, meshy = np.meshgrid(x,y)
meshz = meshx + meshy

axis.plot_surface(meshx,meshy, meshz)
plt.show()
```
