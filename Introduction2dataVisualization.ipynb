```python
import matplotlib.pyplot as plt

print(plt.style.available)
plt.style.use("seaborn")

x = [1,2,3,4,5,6,7]
y = [20,30,40,18,60,70,80]
y2 = [32,49,33,55,66,79,60]


plt.plot(x, y, color="red", linestyle="--")
plt.plot(x, y2, color="green", linestyle="-", marker="p")

plt.title("Muffins sold over the week")
plt.xlabel("Day of the week")
plt.ylabel("Number of Muffins Sold")

plt.legend(["Sales 1", "Sales 2"])

plt.grid(False)
plt.show()
```
