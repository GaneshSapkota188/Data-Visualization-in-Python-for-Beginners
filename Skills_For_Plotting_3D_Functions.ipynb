```python

import matplotlib.pyplot as plt
import numpy as np
axis = plt.axes(projection="3d")

x= np.arange(0,50)
y = np.arange(0,50)
z = x**y

axis.plot(x,y,z, linewidth=3)
axis.set_title("Function")
axis.set_xlabel("X values")
axis.set_ylabel("Y values")
axis.set_zlabel("Z Values")
plt.show()
```
