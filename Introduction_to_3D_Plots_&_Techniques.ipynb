```python
import matplotlib.pyplot as plt
import numpy as np

axis  = plt.axes(projection="3d")

x = np.random.randint(0,100,400)
y = np.random.randint(0,100,400)
z = np.random.randint(0,100,400)
c = np.random.randint(0,10,400)

axis.scatter(x,y,z, s=5, c=c, alpha = 0.9)
plt.show()
```
